```python
def topic_plot2(topicnum=0, title0='', dataset=df2, ylabel='Average Weight', xlabel='Year', xlim1=1740, xlim2=1840):
  mean=df2[topicnum].mean()[xlim1:xlim2]
  std=df2[topicnum].std()[xlim1:xlim2]
#  min=df2[0].min()
#  max=df2[0].max()
  title='Topic '+str(topicnum)+': '+title0
  mean.plot(ylabel=ylabel, xlabel=xlabel, title=title)
  x=mean.index.to_numpy()
  plt.fill_between(x,mean-std,mean+std,color='gray',alpha=0.5)
```
